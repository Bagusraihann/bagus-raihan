```python
distribusi_dan = pd.read_csv('http://data.bandung.go.id/dataset/139c2a1f-9947-4874-82ff-e7dfdc1da98c/resource/9ead1d36-0424-41cd-af30-8ed274bd0468/download/distribusi-dan-kepadatan-penduduk-menurut-kecamatan-di-kota-bandung-2017.csv')
```

```python
path = ('http://data.bandung.go.id/dataset/139c2a1f-9947-4874-82ff-e7dfdc1da98c/resource/9ead1d36-0424-41cd-af30-8ed274bd0468/download/distribusi-dan-kepadatan-penduduk-menurut-kecamatan-di-kota-bandung-2017')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,20))
plt.plot(distribusi_dan['Kepadatan Penduduk (ribu jiwa per km2)'],label = 'Kepadatan Penduduk (dalam ribuan jiwa per km2)')
plt.plot(distribusi_dan['Persentase Penduduk'],label = 'Persentase Penduduk')
plt.xlabel('Data Kecamatan')
plt.ylabel('Nilai Angka')
plt.title('Distribusi dan Kepadatan Penduduk Menurut Kecamatan di Kota Bandung Tahun 2017')
plt.legend()
plt.show()
```
